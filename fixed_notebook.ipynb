```python
from sentence_transformers import SentenceTransformer
```

```python
#load the embedding model
model = SentenceTransformer("intfloat/e5-small-v2")
```

```python
#data
sentences = ["ml learns patterns from data",
             "embedding convert text into numerical vectors",
             "hello!, How are you?",
             "How was your day",
             "transformer are used in llm"]
```

```python
#adding prefix for e5 model
sentences = ["passage: " + s for s in sentences]

embeddings = model.encode(sentences)
```

```python
#taking query input
query = 'what is my name'
query_embedding = model.encode(["query: " + query])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#calculating similarity
score = cosine_similarity(query_embedding, embeddings)[0]
```

```python
#retrieving most similar answer
best_id = score.argmax()
print("Query:", query)
print("Best match:0:", sentences[best_id])
```
